```python
import numpy as np
import time

# Simulate 100 MRI scans (256x256) with random intensities (0-5000)
mriscans = np.random.randint(0, 5000, size=(100, 256, 256))
print ("MRI scans intensities : \n", mriscans)

# Vectorized normalization per scan
normalizedmri = (mriscans - mriscans.min(axis=(1,2), keepdims=True)) / (mriscans.max(axis=(1,2), keepdims=True) - mriscans.min(axis=(1,2), keepdims=True))
print(" MRI normalize intensities : \n", normalizedmri)
#- What it does: Finds the minimum pixel value*for each scan- axis=(1,2) : Compute min across height (axis 1) AND width (axis 2)- keepdims=True: Preserves dimensions as `(100, 1, 1)` instead of `(100,)`

# Check min/max of the normalized data (should be ~0 and ~1)
print("Global min:", normalizedmri.min())  # Expected: ~0.0
print("Global max:", normalizedmri.max())  # Expected: ~1.0

# Check min/max for a single scan (e.g., scan 0)
print("Scan min:", normalizedmri[0].min())  # Should be 0.0
print("Scan max:", normalizedmri[0].max())  # Should be 1.0
print("Confirms that the normalized MRI array retains the original shape :\n ", normalizedmri.shape)
```
